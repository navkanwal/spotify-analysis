```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv(r"C:\Users\navka\Downloads\archive (1)\tracks.csv")
df_tracks.head()
```

```python
#checking whether or not there is any null value present 
#then we will find the sum of all null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#printing least popular 10 songs 

sort_df = df_tracks.sort_values('popularity',ascending = True).head(10)
sort_df
```

```python
df_tracks.describe().transpose()
```

```python
#sorting according to the popularity and printing most popular ones first

most_popular = df_tracks.query('popularity>90',inplace = False).sort_values('popularity',ascending = False)
most_popular[:10]
```

```python
df_tracks.set_index("release_date",inplace = True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head(10)
```

```python
#getting attributes by knowing the row number
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x : round(x/1000))
df_tracks.drop("duration_ms",inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis =1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g" ,vmin=-1, vmax=1 , center=0, cmap="crest",linewidths=1 ,linecolor = "Black" )
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
sample_df
```

```python
plt.figure(figsize=(10,7))
sns.regplot(data =sample_df,y = "loudness",x ="energy",scatter_kws={"color": "black"}, line_kws={"color": "red"}).set(title = "Loudness Vs Energy correlation")
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data = sample_df,y= "popularity",x ="acousticness",scatter_kws={"color": "b"}, line_kws={"color": "c"}).set(title = "Popularity Vs Acousticness correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title = "Number of songs per year")
```

```python
#Bar Graph Plot
total_dr = df_tracks.duration
fig_dims = (18,7)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x= years,y = total_dr,ax=ax,errwidth = False).set(title = "Years vs Duration")
plt.xticks(rotation = 90)
```

```python
total_dr = df_tracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years ,y = total_dr,ax =ax).set(title = "Year vs Duration")
plt.xticks(rotation = 60)
```
